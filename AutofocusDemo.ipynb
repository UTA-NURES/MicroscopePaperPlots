```python
import numpy as np
import matplotlib as mpl
import pylab as plt
import os
from scipy.stats import kurtosis
from Image import *
%matplotlib inline
```

```python

dats_rough=[]    
kurtosis_rough=[]
for file in sorted(os.listdir("/Users/ben.jones/Documents/Data/MicroscopePaper/FocusDemo_3stage/FocusDemo_3.4_to_3.6/"))[1:]:
    dat=np.loadtxt("/Users/ben.jones/Documents/Data/MicroscopePaper/FocusDemo_3stage/FocusDemo_3.4_to_3.6/"+file)
    kurtosis_rough.append(kurtosis(np.ndarray.flatten(dat)))
    dats_rough.append(dat)
    
kurtosis_fine=[]
for file in sorted(os.listdir("/Users/ben.jones/Documents/Data/MicroscopePaper/FocusDemo_3stage/FocusDemoFiner_3.5_to_3.52/"))[1:]:
    dat=np.loadtxt("/Users/ben.jones/Documents/Data/MicroscopePaper/FocusDemo_3stage/FocusDemoFiner_3.5_to_3.52/"+file)
    kurtosis_fine.append(kurtosis(np.ndarray.flatten(dat)))

dats_finer=[]    
kurtosis_finer=[]
for file in sorted(os.listdir("/Users/ben.jones/Documents/Data/MicroscopePaper/FocusDemo_3stage/FocusDemoEvenFiner_3.51_to_3.514/"))[1:]:
    dat=np.loadtxt("/Users/ben.jones/Documents/Data/MicroscopePaper/FocusDemo_3stage/FocusDemoEvenFiner_3.51_to_3.514/"+file)
    kurtosis_finer.append(kurtosis(np.ndarray.flatten(dat)))
    dats_finer.append(dat)
```

```python
FocusPointRough=np.argmax(kurtosis_rough)
stepsize_rough=0.002
X_rough=(range(0,len(kurtosis_rough))-FocusPointRough)*stepsize_rough

FocusPointFiner=np.argmax(kurtosis_finer)
stepsize_finer=0.0001
X_finer=(range(0,len(kurtosis_finer))-FocusPointFiner)*stepsize_finer
```

```python
plt.figure(figsize=(5,5),dpi=150)
plt.plot(X_rough*1000,kurtosis_rough,'o-',color='DarkRed')

plt.xlim(-220,150)
plt.xlabel("Distance from focus (micron)")
plt.ylabel("Autofocus Metric")
plt.savefig("Plots/Autofocus_rough.png",bbox_inches='tight',dpi=200)
```

```python
plt.figure(figsize=(5,5),dpi=150)
plt.plot(X_finer*1000,kurtosis_finer,'o-',color='DarkRed')

plt.xlim(-2,2)
plt.xlabel("Distance from focus (micron)")
plt.ylabel("Autofocus Metric")

plt.ylabel("")
plt.savefig("Plots/Autofocus_fine.png",bbox_inches='tight',dpi=200)
```
